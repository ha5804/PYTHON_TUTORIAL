```python
class Calculate:
    def __init__(self, fs, se):#생성자
        #__init__은 변수관리 수월.
        self.first = fs #self.안에 저장.
        self.second = se

    def add(self):
        result = self.first + self.second
        return 'summation : %d' % result
    
    def multiply(self):
        result = self.first * self.second
        return 'multyply : %d' % result
```

```python
myinstance = Calculate(5, 6)
print(myinstance.add())
print(myinstance.multiply())
```

class 의 상속은, 추가적인 정보가 필요할때, (사람) -> (생일), 사용한다.
이때 상속되는 속성은 `super().__init__`을 통해 생성해야한다.

```python
class Student(Person):
    pass

x = Student("John", "Cho")
x.printname()

class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname,lname)
        self.graduationyear = year
    def printall(self):
        print(self.firstname, self.lastname, self.graduationyear)
```

method overloading은 함수의 이름 중복충돌문제이다. 

```python
def product(a, b):
    p = a * b
    print(p)

def product(a, b, c):
    p = a * b * c
    print(p)

product(4, 5)
```

*arg는 argument의 갯수를 유연하게 넣을 수 있다.

```python
def add(datatype, *args):

    # if datatype is int
    # initialize answer as 0
    if datatype == 'int':
        answer = 0
 
    # if datatype is str
    # initialize answer as ''
    if datatype == 'str':
        answer = ''
 
    # Traverse through the arguments
    for x in args:
 
        # This will do addition if the
        # arguments are int. Or concatenation
        # if the arguments are str
        answer = answer + x
 
    print(answer)
    
add('int', 5, 6)
 
add('str', 'Hi ', 'Geeks')
```

*arg는 고정숫자 뒤에 써야한다. 언제끝날지모르기때문. **karg는 dict형태로 들어간다.

```python
def foo(a, *arg):
    return a, arg

foo(1,2,3)
foo(1,2,3,4)

def foo(a, c, d= 4, *arg, **karg):
    return a, c, d, arg, karg

foo(1,2)
foo(1,2,3,4,5)
foo(1,2,h=2, f=3, d=4, j=3)
```
