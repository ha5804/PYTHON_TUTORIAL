# CSV_DATA & MATPLOT

- DATA_READ

    (1): numpy 이용 하여 csv 데이터 읽어오기

        - np.genfromtxt("filename.csv", delimiter = ",")

    - data 파일에 저장된 point1.csv파일을 읽고 각 행을 x , y 로 구분해보자.

```python
filename = "./point1.csv"
#./는 현재 디렉토리를 의미한다. 모든 경로를 작성할 필요없이 같은 폴더내부 존재하면 위처럼 작성
```

- 경로를 저장했으면 csv파일의 데이터를 처리해보자.

    (csv 파일 내부는 , 를 기준으로 데이터가 나누어져 있다.)

    (csv 파일은 2행으로 나누어져 있다.)

```python
import numpy as np
import csv

point = np.genfromtxt(filename , delimiter = ',')
print(point.shape)
```

- point에 저장된 값이 2행 100열 임을 알 수 있다.

  각 행을 x , y로 표현해보자

```python
x = point[0]
y = point[1]
#2D 행렬기준 슬라이싱은 기본적으로 행을 선택한다.
```

- 처리한 csv 데이터를 plot 해보자.

  (1):x, y축의 글자 크기 수정

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (6, 6))
#figure은 plot사이즈를 결정한다.
plt.plot(x, y, color = 'red', linestyle = ":")
plt.xticks(size = 8)
plt.yticks(size = 8)
plt.show()
```

### 3차원데이터를 불러와 표면을 plot해보자!

(1) surface를 Plot할때는 각 1D배열을 2D배열로 grid해야한다.

    - ax.plot_surface


```python
from matplotlib import cm
#cm을 사용할수있게 하는 라이브러리
filename_3D_Surface = "./point3.csv"
point = np.genfromtxt(filename_3D_Surface, delimiter = ",")
x = point[0]
y = point[1]
z = point[2]
grid_x = x.reshape(100, 100)
grid_y = y.reshape(100, 100)
grid_z = z.reshape(100, 100)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(grid_x, grid_y, grid_z, cmap = cm.coolwarm)
```

# Image 처리

(1)Image를 불러오자.

    - import matplotlib.image as mpimg

```python
import matplotlib.image as mpimg
image = mpimg.imread("./image1_gray.png")
#img의 경로를 미리 설정 해줘야한다.

plt.imshow(image, cmap = 'gray')
plt.axis('off')
```

(2) 이미지를 미분해보자

    - 불러온 이미지를 이차원 행렬로 변경하자

```python
print(image.shape)
#이미지 한장이 600, 800 ,4 라는것은 rgb와 투명도까지 총 4개의 채널이 존재한다는 의미.
image = image[:, :, 0]
#이미지 처리를 위해 2차원 배열로 변경하기
print(image.shape)
```

(3) 미분을 수행해보자

    -행방향 미분, 열방향 미분을 모두 고려해야한다.
     아래에선 행방향으로 미분을 해보자.

```python
h , w = image.shape
#가로 ,세로를 이미지의 shape을 튜플형식으로 받아 저장하자.

df_dy_row = np.zeros((h, w))
#h, w 크기만큼의 행방향으로의 연산자를 0 으로 만들자.
for y in range(h-1):
    for x in range(w):
        df_dy_row[y][x] = image[y+1][x] - image[y][x]
        
df_dy_row[-1, :] = 0

plt.imshow(df_dy_row, cmap = 'gray')
#cmap을 grayscale로 바꾸는이유는 이미지의 밝기변화를 정확하게 파악하기 위해서이다.
```

    -열방향으로 미분을 진행해보자

```python
h, w = image.shape

df_dx_col = np.zeros((h, w))
for y in range(h):
    for x in range(w - 1):
        df_dx_col[y][x] = image[y][x+1] - image[y][x]

# 마지막 열 (경계)은 Neumann 조건 적용
df_dx_col[:, -1] = 0

plt.imshow(df_dx_col, cmap='gray')
plt.axis('off')
```

    - 미분을 진행하여 이미지의 특징을 찾는법을 확인하였다. 효과적인 방법을 찾기위해 

      Laplacian, Gradient norm을 계산해보자.
